```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

df = pd.read_csv('train.csv')
df
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Cabin'], axis=1)
df = df.drop(['Name'], axis=1)
df = df.drop(['Ticket'], axis=1)
#df = df.drop(['Fare'], axis=1)
df = df.drop(['PassengerId'], axis=1)
df
```

```python
df["Sex"].replace("male", 0, inplace=True)
df["Sex"].replace("female", 1, inplace=True)
df
```

```python
df["Embarked"].replace("S", 0, inplace=True)
df["Embarked"].replace("C", 1, inplace=True)
df["Embarked"].replace("Q", 2, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.Age.fillna(df.Age.median(), inplace = True)
df.Embarked.fillna(0, inplace = True)

df.isnull().sum()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', linewidths=0.5)
plt.title('Корреляционная матрица')
plt.show()
```

```python
df = df.drop(['Embarked'], axis=1)
#df = df.drop(['Age'], axis=1)
df = df.drop(['Pclass'], axis=1)
#df = df.drop(['Parch'], axis=1)
#df = df.drop(['SibSp'], axis=1)
df = df.drop(['Fare'], axis=1)

correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', linewidths=0.5)
plt.title('Корреляционная матрица')
plt.show()
```

```python
TargetFun = df["Survived"]
TargetFun
```

```python
for colname in df:
    df[colname] = (df[colname] - min(df[colname])) / (max(df[colname]) - min(df[colname]))

df
```

### Линейная регрессия

```python
#Xtrain, Xtest, Ytrain, Ytest = train_test_split(np.array(df["Survived"]), np.array(df["Fare"]), test_size=0.3, train_size=0.7)
Ytrain, Ytest, Xtrain, Xtest = train_test_split(TargetFun, df.drop(["Survived"], axis=1), test_size=0.3, train_size=0.7)
Ytrain
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, accuracy_score

linreg = LinearRegression()
linreg.fit(Xtrain, Ytrain)

Ypred = linreg.predict(Xtest)
r2 = r2_score(Ytest, Ypred)

r2
```

### Логистическая регрессия

```python
from sklearn.linear_model import LogisticRegression

Ytrain_binary = (Ytrain > .5).astype(int)
Ytest_binary = (Ytest > .5).astype(int)

logreg = LogisticRegression()
logreg.fit(Xtrain, Ytrain_binary)

Ypred = logreg.predict(Xtest)
acc = accuracy_score(Ytest_binary, Ypred)

acc
```

### Модель Лассо

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1)

lasso.fit(Xtrain, Ytrain)
Ypred = lasso.predict(Xtest)
r2 = r2_score(Ytest, Ypred)

r2
```
